```python
import pandas as pd
import sqlalchemy as sql
import datetime

# read password and user to database
credentials_file = r'../credentials.pw'

credentials = pd.read_csv(credentials_file, header=0)
user = credentials['username'].values[0]
pw = credentials['password'].values[0]
```

```python
host="localhost"  # your host
user=user # username
passwd=pw  # password
db="FNL" # name of the database
connect_string = 'mysql+pymysql://%(user)s:%(pw)s@%(host)s:3306/%(db)s'% {"user": user, "pw": pw, "host": host, "db": db}
sql_engine = sql.create_engine(connect_string)
```

```python
def read(sql_engine, table, limit=3):
    """
    Reads last limit entries from table table and returns a dataframe
    """
    query = f"SELECT * FROM {table} ORDER BY id DESC LIMIT {limit}"
    df = pd.read_sql(query, sql_engine)

    return df

def truncate_table(tableName):
    """
    Deletes everything in the table and resets the increment.
    """
    sql_engine.execute(f"DELETE FROM {tableName};")
    sql_engine.execute(f"ALTER TABLE {tableName} AUTO_INCREMENT = 1;")
#     sql_engine.execute(f"SET FOREIGN_KEY_CHECKS = 0;TRUNCATE {tableName};SET FOREIGN_KEY_CHECKS = 1;", multi=True)

def insert_experiment(sql_engine, date, comment, verbose=False):
    """
    Insert an entry into the experiment table.
    """
    table = 'experiment'
    query = f"""INSERT INTO {table} (date, comment) VALUES (\"{date}\", "{comment}\");"""
    if verbose: print(query)
        
    sql_engine.execute(sql.text(query))
    
def insert_tag(sql_engine, tag_name, verbose=False):
    """
    Insert an entry into the tag table.
    """
    table = 'tag'
    query = f"""INSERT INTO {table} (tag_name) VALUES (\"{tag_name}\");"""
    if verbose: print(query)    
    sql_engine.execute(sql.text(query))
    
def insert_user(sql_engine, user_name, comment, verbose=False):
    """
    Insert an entry into the tag table.
    """
    table = 'user'
    query = f"""INSERT INTO {table} (name, comment) VALUES (\"{user_name}\", \"{comment}\");"""
    if verbose: print(query)    
    sql_engine.execute(sql.text(query))
```
